```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Making my plots look nicer
sns.set(style="whitegrid")
```

```python
from sklearn.datasets import load_iris

# Since im using the iris dataset lets load the iris dataset
iris = load_iris(as_frame=True)
df = iris.frame 

# Show first 5 rows
df.head()
```

```python
# My dataset shape
print("Shape of dataset:", df.shape)

# Data types and missing values in the dataset
print("\nData Info:")
print(df.info())

# Checking for missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Summarizing so that i can now start grouping
df.describe()
```

```python
# Lets Group by species and calculate mean of each column
grouped = df.groupby("target").mean()
grouped
```

```python
# Line chart based on sepal length trend
plt.plot(df["sepal length (cm)"])
plt.title("Sepal Length Trend")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

```python
# Bar chart based on average petal length per species
sns.barplot(x="target", y="petal length (cm)", data=df, ci=None)
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
# Histogram of the distribution of sepal width
plt.hist(df["sepal width (cm)"], bins=20, edgecolor="black")
plt.title("Distribution of Sepal Width")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot of sepal length vs petal length
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="target", data=df)
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```
